#Import Dataset

```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/dickyilhami/COVID-19-Indonesia-Case/main/covid_19_indonesia_time_series_all.csv')
```

```python
df1
```

#Data Preparation

```python
df1 = df1.iloc[:-1 , :]
```

```python
df1
```

```python
df1.info()
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df1.info()
```

#Data Cleaning

```python
df1.isnull().mean()*100
```

```python
df1.isna().mean()*100
```

```python
df1 = df1.drop('City or Regency',axis=1)
df1
```

```python
df1['Special Status'].value_counts()
```

```python
df1 = df1.drop('Special Status',axis=1)
df1
```

```python
df1['Growth Factor of New Cases'].head()
```

```python
df1['Growth Factor of New Deaths'].head()
```

```python
df1['Growth Factor of New Cases'] = df1['Growth Factor of New Cases'].fillna(float(0))
df1['Growth Factor of New Deaths'] = df1['Growth Factor of New Deaths'].fillna(float(0))
```

```python
df1[['Province','Island','Time Zone']].head()
```

```python
df1[['Province','Island','Time Zone']] = df1[['Province','Island','Time Zone']].fillna('Unknown')
```

```python
df1['Total Rural Villages'] = df1['Total Rural Villages'].fillna(float(0))
```

```python
df1['Total Urban Villages'] = df1['Total Urban Villages'].fillna(float(0))
```

```python
df1['Total Cities'] = df1['Total Cities'].fillna(float(0))
```

```python
df1
```

```python
df1.isna().mean()*100
```
